```python
def analyze_engagement(likes_list):
    total_likes = 0
    
    
    for likes in likes_list:
        total_likes += likes
    
    print("Total Likes:", total_likes)
    
    if total_likes >= 1000:
        print("Post Status: Viral Post ")
    else:
        print("Post Status: Normal Engagement ")


likes_data = [200, 150, 300, 250, 150]


analyze_engagement(likes_data)
```

```python
def check_medicine_stock(stock):
    print("Medicine Stock:", stock)
    
    if stock < 10:
        print("Status: Low Stock Alert ")
    else:
        print("Status: Stock Sufficient ")


# Example stock value
medicine_stock = 6

# Call function
check_medicine_stock(medicine_stock)
```

```python
def check_rainfall(rainfall_list, required_level):
    total = 0
    
    # Calculate total rainfall using loop
    for rain in rainfall_list:
        total += rain
    
    average = total / len(rainfall_list)
    
    print("Average Rainfall:", average)
    
    if average >= required_level:
        print("Rainfall Status: Adequate Rainfall ")
    else:
        print("Rainfall Status: Inadequate Rainfall ")


# Example rainfall data
rain_data = [70, 80, 65, 75, 70]
required = 70

# Call function
check_rainfall(rain_data, required)
```

```python
def detect_duplicates(usernames):
    unique_users = set(usernames)
    
    if len(unique_users) < len(usernames):
        print("Duplicate Accounts Found: Yes ")
    else:
        print("Duplicate Accounts Found: No ")


# Example usernames list
user_list = ["user1", "user2", "user3", "user1", "user4"]

# Call function
detect_duplicates(user_list)
```

```python
def filter_premium_crops(prices):
    premium = []
    
    
    for price in prices:
        if price > 2000:
            premium.append(price)
    
    print("Premium Crops:", premium)
crop_prices = [1500, 2500, 1800, 3200, 2000]

# Call function
filter_premium_crops(crop_prices)
```

```python
def check_system_health(errors):
    print("Error Count:", errors)
    
    if errors == 0:
        print("System Status: Healthy")
    elif errors <= 5:
        print("System Status: Minor Issues")
    else:
        print("System Status: Critical Issues")


# Example error count
error_count = 7

# Call function
check_system_health(error_count)
```

```python
def check_exam_eligibility(attendance_list):
    present = 0
    
    # Count present days
    for day in attendance_list:
        if day == "P":
            present += 1
    
    total_days = len(attendance_list)
    percentage = (present / total_days) * 100
    
    print("Attendance Percentage:", percentage)
    
    if percentage >= 75:
        print("Exam Eligibility: Eligible")
    else:
        print("Exam Eligibility: Not Eligible")


# Example attendance list
attendance = ["P", "P", "A", "P", "P"]

# Call function
check_exam_eligibility(attendance)
```

```python
def calculate_bill(units):
    bill = 0
    
    
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + ((units - 100) * 5)
    else:
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)
    
    
    if bill < 500:
        status = "Low Usage"
    elif bill <= 1500:
        status = "Moderate Usage"
    else:
        status = "High Usage"
    
    return bill, status

units_consumed = 250

total_bill, usage_status = calculate_bill(units_consumed)

print("Total Bill:", total_bill)
print("Usage Level:", usage_status)
```
