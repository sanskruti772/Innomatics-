# ATM

```python
balance=5000
#min
min_balance=500
#users enters withdrwal amt
withdraw_amount= int(input("enter withdrwal amount:"))
if withdraw_amount %100!=0:
    print("Invalid amount!Enter multiple of 100")
elif balance-withdraw_amount < min_balance:
    print("Transaction failed!:Minimum balance must maintained")
else:
    balance=balance-withdraw_amount
    print("Withdraw successfull")
    print("Remaining balance:",balance)
```

```python
#online shopping kart validation
n= int(input("Enter no of products"))
cart_price =[]
for i in range(n):
    price = float(input(f"enter price of product{i+1}:"))
    cart_price.append(price)
    total=0
    for price in cart_price:
        if price<=0:
            print("Invalid product price")
            break
            total=total+price
        else:
            print("Cart total:",total)
            if total>2000:
                discount=total*0.10
                total=total-discount
                print("discount applied:",discount)
                if total> 500:
                    print("final paymrnt:",total)
                else:
                    print("add some more items in cart for discount")
```
