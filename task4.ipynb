```python
def validate_recharge(amount):
    valid_plans = [199, 299, 399, 599]
    
    if amount < 50:
        print(" Recharge amount must be at least ₹50.")
        return False
    
    if amount not in valid_plans:
        print(" Invalid plan selected.")
        print("Available plans:", valid_plans)
        return False
    
    return True



while True:
    amount = int(input("Enter recharge amount: ₹"))
    
    if validate_recharge(amount):
        print(" Recharge Successful!")
        break
    else:
        print("Please try again.\n")
```

```python
def check_inventory(stock):
    for product, quantity in stock.items():
        if quantity < 15:
            print(product, "→ Reorder Alert (Stock:", quantity, ")")
        else:
            print(product, "→ Stock OK  (Stock:", quantity, ")")


# Dictionary of products and their stock
inventory = {
    "Laptop": 10,
    "Mouse": 25,
    "Keyboard": 12,
    "Monitor": 20
}

# Call function
check_inventory(inventory)
```

```python
def process_result(marks):
    total = 0
    
    # Calculate total using loop
    for m in marks:
        total += m
    
    average = total / len(marks)
    
    # Display result
    print("Average Marks:", average)
    
    if average >= 50:
        return "Pass "
    else:
        return "Fail "


# Example marks list
student_marks = [60, 45, 70, 55]

# Call function
result = process_result(student_marks)
print("Result:", result)
```

```python
def calculate_fare(distance, peak):
    base_fare = 50
    per_km_rate = 12
    
    fare = base_fare + (distance * per_km_rate)
    
    if peak.lower() == "yes":
        fare = fare + (fare * 0.25)   # 25% extra
    
    return fare


# Retry system
while True:
    distance = float(input("Enter distance in km: "))
    peak = input("Is it peak hour? (yes/no): ")
    
    total_fare = calculate_fare(distance, peak)
    print("Total Fare: ₹", total_fare)
    
    retry = input("Do you want to calculate again? (yes/no): ")
    if retry.lower() != "yes":
        print("Thank you for using Cab Fare Estimator ")
        break
```

```python
def check_eligibility(attendance):
    present_count = 0
    
    # Count present days
    for status in attendance:
        if status == "P":
            present_count += 1
    
    total_days = len(attendance)
    percentage = (present_count / total_days) * 100
    
    print("Attendance Percentage:", percentage, "%")
    
    if percentage >= 75:
        return "Eligible "
    else:
        return "Not Eligible "


# Example attendance list
attendance_list = ["P", "A", "P", "P", "A", "P", "P", "P"]

# Call function
result = check_eligibility(attendance_list)
print("Status:", result)
```

```python
def check_password(password):
    has_digit = False
    has_special = False
    special_chars = "@#$"
    
    # Check conditions using loop
    for ch in password:
        if ch.isdigit():
            has_digit = True
        if ch in special_chars:
            has_special = True
    
    # Validate rules
    if len(password) < 8:
        return " Weak Password (Minimum length 8 required)"
    elif not has_digit:
        return " Weak Password (Must contain at least one digit)"
    elif not has_special:
        return " Weak Password (Must contain special character @#$)"
    else:
        return " Strong Password"
pwd = input("Enter your password: ")
result = check_password(pwd)
print(result)
```
